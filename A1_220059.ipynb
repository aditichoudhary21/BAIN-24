Q1. Suppose you have a jar with 4 jelly beans: 2 are strawberry-flavored, 1 is blueberry-flavored, and 1 is cinnamon-flavored. You draw one jelly bean at random from the jar.

    a. What is the sample space for this experiment?

    b. We define event A as the jelly bean drawn is strawberry-flavored and event B as The jelly bean drawn is not cinnamon-flavored. What are the probabilities of events A and B?

    c. Are events A and B mutually exclusive? Why or why not?


Soln 1. a. Sample Space:
   S = {strawberry, strawberry, blueberry, cinnamon}

b. Probability of Event A, P(A) = 2/4 = 0.5

   Probability of Event B, P(B) = 3/4 = 0.75

c. Events A and B are not mutually exclusive because it is possible for an outcome to belong to both events simultaneously.
    For example, if a strawberry-flavored jelly bean is drawn (event A), it is also true that the jelly bean is not cinnamon-flavored (event B).
    Therefore, events A and B can occur together. Mathematically, events A and B are not mutually exclusive if P(A ∩ B) > 0. In this case, P(A ∩ B) = 1/4,
   as there is one jelly bean that is both strawberry-flavored and not cinnamon-flavored.

 Q2. Previously, we defined a **Python function P** to compute the probability of an event using the naive definition of probability. Generalize that function to compute the probability of events when they are
 not all equally likely. Use this new function to compute the probability of events A and B from the previous exercise.

 Hint: you can pass a third argument with the probability of each event.

```python
def P(S, A, probabilities):
    if set(A).issubset(set(S)):
        total_probability = sum(probabilities[key] for key in A)
        return total_probability
    else:
        return 0

# Define the sample space S
S = ['strawberry', 'strawberry', 'blueberry', 'cinnamon']

# Define the probabilities for each outcome/event
probabilities = {'strawberry': 2/4, 'blueberry': 1/4, 'cinnamon': 1/4}

# Define event A as the jelly bean drawn is strawberry-flavored
A = ['strawberry']

# Define event B as the jelly bean drawn is not cinnamon-flavored
B = ['strawberry', 'blueberry']

# Calculate the probability of events A and B using the corrected function P
probability_A = P(S, A, probabilities)
probability_B = P(S, B, probabilities)

# Print the results
print("Probability of event A:", probability_A)
print("Probability of event B:", probability_B)
```

Q3.  Use PreliZ to explore different parameters for the Gaussian distributions. Use the methods `plot_pdf`, `plot_cdf`, and `plot_interactive`. We discussed the probability mass/density functions and the cumulative density function. But there are other ways to represent functions like the percentile point function ppf. Using the `plot_ppf` method of PreliZ, plot the percentile point function for the BetaBinomial and Gaussian distributions. Can you explain how the ppf is related to the cdf and pmf/pdf?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define parameters for the Gaussian distribution
mean = 0
std_dev = 1

# Create a Gaussian distribution object
gaussian_dist = stats.norm(mean, std_dev)

# Generate a range of probabilities for Gaussian distribution
gaussian_probabilities = np.linspace(0.01, 0.99, 100)  # Probabilities from 0.01 to 0.99

# Calculate the PPF values for the Gaussian distribution
gaussian_ppf_values = gaussian_dist.ppf(gaussian_probabilities)

# Plot the PPF for the Gaussian distribution
plt.figure(figsize=(8, 4))
plt.plot(gaussian_probabilities, gaussian_ppf_values)
plt.xlabel('Probability')
plt.ylabel('Percentile Point Function (PPF)')
plt.title('Percentile Point Function (PPF) of Gaussian Distribution')
plt.grid(True)
plt.show()

# Define parameters for the BetaBinomial distribution
n = 10  # Number of trials
a = 2   # Shape parameter alpha
b = 5   # Shape parameter beta

# Create a BetaBinomial distribution object
betabinom_dist = stats.betabinom(n, a, b)

# Generate a range of probabilities for BetaBinomial distribution
betabinomial_probabilities = np.linspace(0.01, 0.99, 100)  # Probabilities from 0.01 to 0.99

# Calculate the PPF values for the BetaBinomial distribution
betabinomial_ppf_values = betabinom_dist.ppf(betabinomial_probabilities)

# Plot the PPF for the BetaBinomial distribution
plt.figure(figsize=(8, 4))
plt.plot(betabinomial_probabilities, betabinomial_ppf_values)
plt.xlabel('Probability')
plt.ylabel('Percentile Point Function (PPF)')
plt.title('Percentile Point Function (PPF) of BetaBinomial Distribution')
plt.grid(True)
plt.show()
```

Q4. Use PreliZ to compute the moments for the SkewNormal distribution for a different combination of parameters. Generate random samples of different sizes,
like 10, 100, and 1,000, and see if you can recover the values of the first two moments (mean and variance) from the samples. Report your obervations.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define parameters for the SkewNormal distribution
a = 0.5  # Skewness parameter
mean = 2  # Mean
std_dev = 1  # Standard deviation

# Create a SkewNormal distribution object
skewnorm_dist = stats.skewnorm(a, loc=mean, scale=std_dev)

# Compute the moments (mean and variance) using PreliZ
mean_preliZ = skewnorm_dist.mean()
variance_preliZ = skewnorm_dist.var()

# Print the moments computed using PreliZ
print("Moments computed using PreliZ:")
print("Mean:", mean_preliZ)
print("Variance:", variance_preliZ)

# Generate random samples of different sizes
sample_sizes = [10, 100, 1000]
for size in sample_sizes:
    # Generate random samples
    samples = skewnorm_dist.rvs(size=size, random_state=42)

    # Compute moments from the samples
    mean_sample = np.mean(samples)
    variance_sample = np.var(samples, ddof=0)  # Use ddof=0 for population variance

    # Print the moments computed from the samples
    print(f"\nSample size: {size}")
    print("Mean (sample):", mean_sample)
    print("Variance (sample):", variance_sample)

    # Plot histogram of the samples
    plt.figure(figsize=(8, 4))
    plt.hist(samples, bins=20, density=True, alpha=0.7, color='blue')
    plt.title(f"Histogram of SkewNormal Samples (Size: {size})")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()
```
